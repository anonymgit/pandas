# Pandas concat()

```python
#pandas provides various facilities for easily combining together Series,Dataframe and Panel objects.
# parameters:
#     objs,ignore_index,axis,keys,join,sort,join_axes
```

```python
import pandas as pd
```

```python
#sereis 1
s1=pd.Series([0,1,2])
s1
```

```python
#series 2
# s2=pd.Series([3,4,5])
# s2
s2=pd.Series([3,4,5,6,7])
s2
```

# objs


```python
pd.concat([s1,s2])
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[5,6,7,8]})
#must have equal index numbers
df1
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],'Name':['E','F','G','H'],'Class':[9,10,11,12]})
#must have equal index numbers
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

# axis


```python
pd.concat([df1,df2],axis=1)  #axis=1 concat data column wise
```

```python
pd.concat([df1,df2],axis=0)
```

# ignore_index


```python
#for proper indexing use ignore_index=true/false
#default false
pd.concat([df1,df2],ignore_index=True)
```

```python
df4=pd.DataFrame({'ID':[5,6,7,8],'Name':['E','F','G','H'],'Class':[9,10,11,12]})
df4
```

```python
df5=pd.DataFrame({'ID':[5,6],'Name':['E','F'],'Class':[9,10]})

df5
```

```python
pd.concat([df4,df5],ignore_index=True)
```

```python
pd.concat([df4,df5],ignore_index=True,axis=1)
```

# join(inner,outer)

```python
pd.concat([df4,df5],axis=1,join='inner')
```

```python
pd.concat([df4,df5],axis=1,join='outer')
```

```python
pd.concat([df4,df5],axis=1,keys=['first df','second df'])
```
