```python
import pandas as pd;
```

```python
titanic=pd.read_csv("D:/titanic.csv")
titanic.head(10)
```

```python
titanic.interpolate()  #it will predict and fill the NaN value (it will fill only those column whose data type is integer)
titanic.head(10)
```

```python
type(titanic.Age[0])
#age column contain float data type lets convert its type
```

```python
titanic.fillna({"Age":0}).head(10)
```

```python
#You can then replace the NaN values with zeros by adding fillna(0), 
#and then perform the conversion to integers using astype(int):

titanic['Age'] = titanic['Age'].fillna(0).astype(int)
```

```python
titanic.head(10)
```

```python
type(titanic.Age[0])  #data type is converted
```

```python
#now we use interpolate on age
titanic.interpolate()
```
